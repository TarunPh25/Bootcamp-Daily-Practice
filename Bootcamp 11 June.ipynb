# Tkinter:
Tkinter is the standard GUI toolkit for Python. It is included with most Python installations and provides a way to create windows, dialogs, buttons, and other GUI components.

# Setting Up Tkinter:
Tkinter is included with Python, so you usually don't need to install anything extra. You can start using it by importing the tkinter module in your Python script.

# Basic Tkinter Components


->Label: Display static text or images.

label = tk.Label(root, text="This is a label")
label.pack()

->Button: Create clickable buttons to trigger actions.

def on_click():
    print("Button clicked!")

button = tk.Button(root, text="Click Me", command=on_click)
button.pack()

->Entry: Allow user input for single-line text.

entry = tk.Entry(root)
entry.pack()

->Frame: Organize widgets within a container.

frame = tk.Frame(root)
frame.pack()

->Check Button: Create checkboxes for boolean options.

check = tk.Checkbutton(root, text="Check me")
check.pack()

->Radio Button: Create radio buttons for mutually exclusive options.

radio1 = tk.Radiobutton(root, text="Option 1", value=1)
radio2 = tk.Radiobutton(root, text="Option 2", value=2)
radio1.pack()
radio2.pack()

->Listbox: Display a list of items for selection.

listbox = tk.Listbox(root)
listbox.insert(1, "Item 1")
listbox.insert(2, "Item 2")
listbox.pack()

->Scrollbar: Add scrollbars to widgets like Listbox.

scrollbar = tk.Scrollbar(root)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

->Menu: Create simple menus for application navigation.

menubar = tk.Menu(root)
filemenu = tk.Menu(menubar, tearoff=0)
filemenu.add_command(label="Open")
filemenu.add_command(label="Save")
filemenu.add_separator()
filemenu.add_command(label="Exit", command=root.quit)
menubar.add_cascade(label="File", menu=filemenu)
root.config(menu=menubar)

->Canvas: Draw shapes, lines, text, and images.

canvas = tk.Canvas(root, width=200, height=200)
canvas.create_line(0, 0, 200, 100)
canvas.pack()


Tkinter Layout Management:

->pack(): Organizes widgets in blocks before placing them in the parent widget.

label = tk.Label(root, text="Packed Label")
label.pack(side=tk.LEFT, padx=5, pady=5)

->grid(): Organizes widgets in a table-like structure.

label1 = tk.Label(root, text="Label 1")
label1.grid(row=0, column=0)
label2 = tk.Label(root, text="Label 2")
label2.grid(row=0, column=1)

->place(): Places widgets at an absolute position.

label = tk.Label(root, text="Placed Label")
label.place(x=50, y=50)
