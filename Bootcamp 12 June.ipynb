```python
from tkinter import *

root = Tk()

#creating a label widget
myLabel = Label(root,text="Hello Tarun!")
#Shoving it onto the screen
myLabel.pack()

root.mainloop()
#Positiong with Tkinter's Grid system:
from tkinter import *
root = Tk()
#creating a label widget
mylabel1 = Label(root,text="Hello Tarun!").grid(row=0, column=0)
mylabel2 = Label(root,text="My name is Tarun").grid( row=1, column=5)
#showing it onto the screen
#mylabel1.grid(row=0, column=0)
#mylabel2.grid(row=1, column=0)
root.mainloop()
# Creating Button
from tkinter import *
root = Tk()

def myclick():
  mylabel1= Label(root,text="Look I click a button!")
  mylabel1.pack()
#for making a button in square and in center from left to right 
#myButton = Button(root,text="Click me!", pady=50, padax=50)
#myButton.pack()
myButton = Button(root,text="Click me!", command=myClick)
myButton.pack()
#for colouring background(red)and forground(blue) of button
#myButton = Button(root,text="Click me!",fg="blue",bg="red")
#myButton.pack()
#for colouring background(white)and forground(blue) of button
#myButton = Button(root,text="Click me!",fg="blue",bg="#ffffff")
#myButton.pack()
#for colouring background(black)and forground(blue) of button
#myButton = Button(root,text="Click me!",fg="blue",bg="#000000")
#myButton.pack()
root.mainloop()

#creating Input Fields
from tkinter import *
root = Tk()
#for seting up width, bg and fg color, borderwidth) 
e = Entry(root, width=50, borderwidth=5, bg="blue", fg="white")
e.pack()
e.insert(0, "enter your name:")
def myclick():
  # concatenate hello text and user enter name
  mylabel1= Label(root,text="Hello! " + e.get())
  mylabel1.pack()

myButton = Button(root,text="Enter your name", command=myClick)
myButton.pack()
root.mainloop()

#Build a Simple calculator app
from tkinter import *

root = Tk()
root.title("Simple Calculator")

e = Entry(root, width=35, borderwidth=5)
e.grid(row=0, column=0, columnspan=3, padx=10, pady=10)

#e.insert(0, "")
#make a function button_click for adding, deletingand find current number :
def button_click(number):
	#e.delete(0, END)
	current = e.get()
	e.delete(0, END)
	e.insert(0, str(current) + str(number))
#for clear all the numbers
def button_clear():
	e.delete(0, END)
#for adding given numbers
def button_add():
	first_number = e.get()
	global f_num
	global math
	math = "addition"
	f_num = int(first_number)
	e.delete(0, END)

def button_equal():
	second_number = e.get()
	e.delete(0, END)
	
	if math == "addition":
		e.insert(0, f_num + int(second_number))

	if math == "subtraction":
		e.insert(0, f_num - int(second_number))

	if math == "multiplication":
		e.insert(0, f_num * int(second_number))

	if math == "division":
		e.insert(0, f_num / int(second_number))

	
#For subtraction of the numbers:
def button_subtract():
	first_number = e.get()
	global f_num
	global math
	math = "subtraction"
	f_num = int(first_number)
	e.delete(0, END)
#For multiplication of the numbers 
def button_multiply():
	first_number = e.get()
	global f_num
	global math
	math = "multiplication"
	f_num = int(first_number)
	e.delete(0, END)
#For division of the numbers 
def button_divide():
	first_number = e.get()
	global f_num
	global math
	math = "division"
	f_num = int(first_number)
	e.delete(0, END)


# Define Buttons

button_1 = Button(root, text="1", padx=40, pady=20, command=lambda: button_click(1))
button_2 = Button(root, text="2", padx=40, pady=20, command=lambda: button_click(2))
button_3 = Button(root, text="3", padx=40, pady=20, command=lambda: button_click(3))
button_4 = Button(root, text="4", padx=40, pady=20, command=lambda: button_click(4))
button_5 = Button(root, text="5", padx=40, pady=20, command=lambda: button_click(5))
button_6 = Button(root, text="6", padx=40, pady=20, command=lambda: button_click(6))
button_7 = Button(root, text="7", padx=40, pady=20, command=lambda: button_click(7))
button_8 = Button(root, text="8", padx=40, pady=20, command=lambda: button_click(8))
button_9 = Button(root, text="9", padx=40, pady=20, command=lambda: button_click(9))
button_0 = Button(root, text="0", padx=40, pady=20, command=lambda: button_click(0))
button_add = Button(root, text="+", padx=39, pady=20, command=button_add)
button_equal = Button(root, text="=", padx=91, pady=20, command=button_equal)
button_clear = Button(root, text="Clear", padx=79, pady=20, command=button_clear)

button_subtract = Button(root, text="-", padx=41, pady=20, command=button_subtract)
button_multiply = Button(root, text="*", padx=40, pady=20, command=button_multiply)
button_divide = Button(root, text="/", padx=41, pady=20, command=button_divide)

# Put the buttons on the screen

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)

button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)

button_0.grid(row=4, column=0)
button_clear.grid(row=4, column=1, columnspan=2)
button_add.grid(row=5, column=0)
button_equal.grid(row=5, column=1, columnspan=2)

button_subtract.grid(row=6, column=0)
button_multiply.grid(row=6, column=1)
button_divide.grid(row=6, column=2)







root.mainloop()
```
